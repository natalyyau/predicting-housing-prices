```python
import numpy as np

# Simulate 10 coin tosses (1 = heads, 0 = tails)
tosses = np.random.randint(0, 2, 10) # 10 numbers between 0 and 1 (2 is not included)

print("Coin toss results:", tosses)
print("Number of heads:", np.sum(tosses))
print("Proportion of heads:", np.mean(tosses))
```
