## Sorting

In pandas, there are two main ways to sort a DataFrame:


#### `sort_index()`
Sorts by row or column labels.  
Default is along rows (`axis=0`), but you can also sort columns with `axis=1`.

 
`df.sort_index()`  sorts rows by index  
`df.sort_index(axis=1)`  sorts columns by names  



#### `sort_values()`
Sorts by the values in one or more columns.  
Use the `by` parameter to choose the column(s).

 
`df.sort_values(by="A")`  sorts rows by column A  
`df.sort_values(by=["B","C"])`  sorts rows by B, then by C  


````python
df
````

````python
df.sort_index(axis=0, ascending=True)
````

````python
df.sort_values(by=["A", "B", 'C'], ascending=[True,False, True])
````

## Joining two data frames

- Use `concat()` when you want to stack DataFrames together (row-wise or column-wise), without needing a key match.  
- Use `join()` when you mostly join by index (and want simpler syntax).
- Use `merge()` when you join on columns (like SQL joins).

#### `concat()`  combines DataFrames by stacking along an axis

- **Row-wise concatenation (default, `axis=0`):** stack DataFrames on top of each other.  
- **Column-wise concatenation (`axis=1`):** place DataFrames side by side.  

- **Outer join (default):** keep all labels from both objects.  
- **Inner join:** keep only labels present in both (use `join="inner"`).  

- Use `ignore_index=True` to reset the index in the result.  
- Unlike `join()` or `merge()`, `concat()` does not require matching keys. It just aligns on index/columns and fills gaps with `NaN`.  

````python
df1 = pd.DataFrame({"id": [1, 2, 3], "A": [1, 2, 9], "B": [10, 20, 90]}, index=["a", "b", "c"])
df2 = pd.DataFrame({"id": [1, 3, 5], "A2": [3, 4, -1], "B2": [10, 40, 0]}, index=["b", "c", "d"])

# Row-wise concatenation
pd.concat([df1, df2])
````

````python
# Column-wise concatenation
pd.concat([df1, df2], axis=1)
````

````python
# Inner join
pd.concat([df1, df2], join='inner')
````

````python
pd.concat([df1, df2], keys=["df1", "df2"])
````

#### `join()`  joins on **index** by default

- **Left join (default):** keep all indices from the left DataFrame.  
- **Right join:** keep all indices from the right DataFrame.  
- **Inner join:** keep only the indices present in both DataFrames.  
- **Outer join:** keep all indices from both DataFrames.  

- By default, joins are done on the **index**.  
- You can also join on a column by setting the other DataFrame's index.  

- If columns have the same name in both DataFrames, use `lsuffix` and `rsuffix` to distinguish them.  

````python
df1 = pd.DataFrame({"id": [1, 2, 3], "A": [1, 2, 9], "B": [10, 20, 90]}, index=["a", "b", "c"])
df2 = pd.DataFrame({"id2": [1, 3, 5], "A2": [3, 4, -1], "B2": [10, 40, 0]}, index=["b", "c", "d"])

df1.join(df2)
````

````python
students = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Major": ["CS", "Math", "DS"]
}, index=["s1", "s2", "s3"])

scores = pd.DataFrame({
    "Course": ["Fundamentals of Data Science", "Data Mining", "Machine Learning"],
    "Score": [85, 92, 78]
}, index=["s1", "s2", "s4"]) 

display(students, scores)
````

````python
# Default join (left join): keep all indices from the left DataFrame
students.join(scores)
````

````python
# Right join: keep all indices from the right DataFrame
students.join(scores, how='right')
````

````python
# Inner join: join on intersection of indices
students.join(scores, how="inner")
````

````python
# Outer join: join on union of indices
students.join(scores, how="outer")
````

````python
# Joining on a specific column instead of the index
students.join(scores, on="Major")
````

````python
students = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Major": ["CS", "Math", "DS"]
}, index=["s1", "s2", "s3"])

scores = pd.DataFrame({
    "Course": ["Fundamentals of Data Science", "Data Mining", "Machine Learning"],
    "Score": [85, 92, 78],
    "Major": ["CS", "Math", "Biology"]
}, index=["s1", "s2", "s4"]) 

display(students, scores)
````

#### `merge()`  joins on **common column names** by default

- **Inner merge (default):** keep only the rows where keys (= values of columns with the same names) match in both DataFrames.  
- **Outer merge:** for each row:
    - if keys (= values of columns with the same names) match in both DataFrames, merges rows
    - if keys don't match, adds necessary NaNs to each of rows, adds both rows  
- **Left merge:** keep all rows from the left DataFrame, matching with the right where possible.  
- **Right merge:** keep all rows from the right DataFrame, matching with the left where possible.  

- Joins are performed on one or more **columns (keys)**.  
- You can also merge on the *index* using `left_index=True` and `right_index=True`.  

- If non-key columns have the *same name* in both DataFrames, pandas adds suffixes (`_x`, `_y` by default).  
  You can set custom suffixes with the `suffixes` parameter.  

````python
df1 = pd.DataFrame({"id": [1, 3, 3], "A": [3, 2, 9], "B": [10, 20, 90]})
df2 = pd.DataFrame({"id": [1, 3, 5], "A": [3, 4, -1], "B2": [10, 40, 0]})
display(df1, df2)
````

````python
# Inner merge (default): keep only the rows where keys (= values of columns with the same names) match in both DataFrames.

df1.merge(df2)
````

````python
# Outer merge: for each row:
#   if keys (= values of columns with the same names) match in both DataFrames, merges rows
#   if keys don't match, adds necessary NaNs to each of rows, adds both rows
df1.merge(df2, how='outer')
````

````python
# Left merge: keep all rows from the left DataFrame, matching with the right where possible.  
df1.merge(df2, how='left')
````

````python
# Right merge: keep all rows from the right DataFrame, matching with the right where possible.  
df1.merge(df2, how='right')
````

````python
df1.merge(df2, left_index=True, right_index=True, suffixes=("_1", "_2"))
# df1.merge(df2, left_index=True, right_on='A')
````

## Grouping data

`.groupby('some_category')` groups all rows having the same attributes (as specified in by parameter) into separate data frames.

After the groupby, you need an aggregate function to summarize data in each subframe.

You may do it in 2 ways:

````python
df
````

````python
# In each subframe, take the `price` column and summarize it
# using the `sum` function from the result
df.groupby(['some_category'])['price'].sum()
````

````python
# In each subframe, apply the `sum` function to all numeric
# columns then extract the `price` column
df.groupby(['some_category']).sum()['price']
````

````python
df.groupby("some_category").agg({
    "new": "sum",
    "price": "median"
})
````

## Pivot Tables

Pivot tables allow you to summarize data by reorganizing it into a table where:
- Rows represent one set of keys,
- Columns represent another set of keys,
- Cell values are an aggregated function (like sum, mean, count, etc.).



### Creating a Pivot Table

Use `pd.pivot_table()`:

```python
pd.pivot_table(
    data=df,
    values="column_to_aggregate",
    index="column_for_rows",
    columns="column_for_columns",
    aggfunc="function"
)


````python
df
````

````python
df.pivot_table(index='some_category', columns='one more category',values='price')
````

````python
df.pivot_table(index='some_category', columns='one more category', values='price', aggfunc='sum')
````

````python
df.pivot_table(index='some_category', columns='one more category', values='price', aggfunc=['sum', 'mean'])
````
