```python
import numpy as np
import pandas as pd

A= np.arange(1,21)
A = A.reshape(5,4)
A
```

```python
row_sums = np.sum(A, axis=1)   # sum of rows in a list
column_sums = np.sum(A, axis=0) # sum of colunns in a list

print(row_sums, column_sums)

np.sum(row_sums), np.sum(column_sums)  # sum of rows (total), sum of columns(total)
```

```python
def row_vector(list):  #take a list as a function and return in a row vector
    return np.array(list).reshape(1,len(list))

s = row_vector([1,2,3])
print(s)
```

```python
def column_vector(list):  #take a list as a function and return in a column vector
    return np.array(list).reshape(len(list),1)
```
